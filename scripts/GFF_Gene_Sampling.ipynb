```python
import os
import re
import random
import time
```

```python
start_time = time.time()
gff_file_path = '/Users/mohanavenkataphaneendrareddyalla/Downloads/Yeast_R64Version_GenomeAnnotation.gff'
```

```python
genes_info = []

with open(gff_file_path, 'r') as file:
    for line in file:
        if line.startswith("#"):
            continue  
        fields = line.strip().split("\t")
        if fields[2] == "gene":  # Chekcing'gene'
            chrom_id = fields[0]
            start_pos = fields[3]
            end_pos = fields[4]
            gene_id_match = re.search(r"ID=([^;]+)", fields[8])
            if gene_id_match:
                gene_id = gene_id_match.group(1)
                genes_info.append((gene_id, chrom_id, start_pos, end_pos))
```

```python
import os
os.makedirs('My_Temp_Folder')
```

```python
for i in range(1, 11):
    selected_genes = random.sample(genes_info, 100)
    temp_file_path = f'My_Temp_Folder/My_Temp_File_{i}.txt'
    with open(temp_file_path, 'w') as temp_file:
        for gene_info in selected_genes:
            temp_file.write("\t".join(gene_info) + "\n")
```

```python
end_time = time.time()
print(f"Program execution time: {end_time - start_time} seconds")
```
