```python
def read_gene_expression_file(file_name):
   
    gene_expression_dict = {}
    
    # Open the file for reading
    with open('/Users/mohanavenkataphaneendrareddyalla/Desktop/Gene_Expression.txt', 'r') as file:
        # Skip the header row
        next(file)
        
        # Read each line in the file
        for line in file:
            # Split the line into columns
            columns = line.strip().split('\t')  # Assuming the data is tab-separated
            gene_name = columns[0]  # First column is the gene name
            expression_levels = list(map(float, columns[1:]))  # Convert expression levels to floats
            gene_expression_dict[gene_name] = expression_levels  # Store in dictionary
    
    return gene_expression_dict
```

```python
def calculate_statistics(gene_expression_dict):
    
    statistics_dict = {}
    
    for gene, levels in gene_expression_dict.items():
        min_level = min(levels)  # Calculate minimum
        max_level = max(levels)  # Calculate maximum
        avg_level = sum(levels) / len(levels)  # Calculate average
        statistics_dict[gene] = (min_level, max_level, avg_level)  # Store in dictionary
    
    return statistics_dict
```

```python
def write_results_to_file(statistics_dict, output_file_name):
   
    with open(output_file_name, 'w') as file:
        # Write each gene and its statistics to the file
        for gene, (min_val, max_val, avg_val) in statistics_dict.items():
            file.write(f"{gene}\t{min_val}\t{max_val}\t{avg_val}\n")  # Tab-separated values
```

```python
# Define the input and output file paths
input_file_name = '/Users/mohanavenkataphaneendrareddyalla/Desktop/Gene_Expression.txt'
output_file_name = '/Users/mohanavenkataphaneendrareddyalla/Desktop/Gene_Statistics.txt'

# Step 1: Read the gene expression data
gene_expression_data = read_gene_expression_file(input_file_name)

# Step 2: Calculate statistics
gene_statistics = calculate_statistics(gene_expression_data)

# Step 3: Write the results to a file
write_results_to_file(gene_statistics, output_file_name)

# Confirm that the output file was created
print(f"Results written to {output_file_name}")
```
