```python

import math
```

```python
# Step 1
gene_expression = {}

print("Step 1: Reading the Gene Expression data")
with open('/Users/mohanavenkataphaneendrareddyalla/Downloads/Gene_Expression.txt', 'r') as f:
    # Ignore the header
    header = f.readline().strip().split()
    print("Header detected: ", header)
    
    # Read each gene and its expression levels
    for line in f:
        parts = line.strip().split()
        gene_name = parts[0]
        expression_levels = list(map(float, parts[1:]))
        gene_expression[gene_name] = expression_levels
        print(gene_name , expression_levels)

print("All gene expression data loaded into the dictionary.")
```

```python
# Step 2: Calculate min, max, average, sum, and variance
def calculate_statistics(expression_levels):
    n = len(expression_levels)
    total = sum(expression_levels)
    mean = total / n
    variance = sum((x - mean) ** 2 for x in expression_levels) / (n - 1) if n > 1 else 0
    return min(expression_levels), max(expression_levels), mean, total, variance

print("\nStep 2: Calculating statistics (min, max, average, sum, variance) for each gene.")
# Print statistics for each gene
for gene, levels in gene_expression.items():
    min_val, max_val, mean_val, sum_val, variance_val = calculate_statistics(levels)
    print(f"\nGene: {gene}")
    print(f"Min: {min_val}, Max: {max_val}, Mean: {mean_val}, Sum: {sum_val}, Variance: {variance_val}")

print("\nStep 2 Complete: Statistics calculated for all genes.")
```

```python
# Step 3: Sort genes by average expression levels and print top 10
print("\nStep 3: Sorting genes by average expression levels.")
sorted_by_mean = sorted(gene_expression.items(), key=lambda item: sum(item[1])/len(item[1]), reverse=True)

print("Top 10 genes by average expression levels:")
i = 0
while i < len(sorted_by_mean) and i < 10:
    gene, levels = sorted_by_mean[i]
    mean_val = sum(levels) / len(levels)
    print(f"{i+1}. {gene}: {mean_val}")
    i += 1
print("\nStep 3 Complete: Top 10 genes by average expression printed.")
```

```python
# Step 4: Sort genes by variance and print top 10
print("\nStep 4: Sorting genes by variance.")
# Sort genes by variance
sorted_by_variance = sorted(gene_expression.items(), key=lambda item: calculate_statistics(item[1])[4], reverse=True)

print("Top 10 genes by variance:")
i = 0
while i < len(sorted_by_variance) and i < 10:
    gene, levels = sorted_by_variance[i]
    variance_val = calculate_statistics(levels)[4]
    
    # Skip genes with variance below a certain threshold
    if variance_val < 1:  # Change this threshold as needed
        i += 1
        continue
    
    print(f"{i + 1}. {gene}: {variance_val}")
    i += 1

print("\nStep 4 Complete: Top 10 genes by variance printed.")
```

```python
# Step 5: Calculate log2 fold changes and write to file
print("\nStep 5: Calculating log2 fold changes and writing to 'Gene_Expression_log2FC.txt'.")
gene_names = list(gene_expression.keys())

with open('Gene_Expression_log2FC.txt', 'w') as f_out:
    # Write header
    f_out.write('\t' + '\t'.join(gene_names) + '\n')
    
    for i, gene1 in enumerate(gene_names):
        f_out.write(gene1)
        for j, gene2 in enumerate(gene_names):
            mean1 = sum(gene_expression[gene1]) / len(gene_expression[gene1])
            mean2 = sum(gene_expression[gene2]) / len(gene_expression[gene2])
            log2_fc = math.log2(mean1 / mean2) if mean2 != 0 else 'inf'
            f_out.write(f'\t{log2_fc:.2f}' if log2_fc != 'inf' else '\tinf')
        f_out.write('\n')

print("\nStep 5 Complete: Log2 fold changes written to 'Gene_Expression_log2FC.txt'.")
```
