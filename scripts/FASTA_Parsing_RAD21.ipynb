```python
from Bio import SeqIO

# File path to your FASTA file
file_path = "/Users/mohanavenkataphaneendrareddyalla/Desktop/Adv_pro_molecular_bio/Homework_3/RAD21_Protein_Sequences.txt"

# Create an empty dictionary to store the sequences
sequences = {}

# Use SeqIO to parse the FASTA file and store the data in the dictionary
for record in SeqIO.parse(file_path, "fasta"):
    key = ">" + record.id  # Create the key by concatenating ">" and the sequence ID
    value = record.seq._data  # Directly access the sequence as a string
    sequences[key]= value

# Print the parsed data
print(sequences)

```

```python
# Create a list of the lengths of each sequence
lengths = [len(seq) for seq in sequences.values()]

# Print the unsorted list of gene sequence lengths
print("Unsorted Gene Lengths:\n", lengths)
```

```python
# Sort the lengths list in ascending order
sorted_lengths = sorted(lengths)

# Print the sorted list of gene sequence lengths
print("Sorted Gene Lengths:\n", sorted_lengths)
```

```python
# Initialize variables to store RAD21_MOUSE gi ID and sequence
rad21_mouse_id = None
rad21_mouse_sequence = None

# Search for the gene "RAD21_MOUSE" in the dictionary
for header, seq in sequences.items():
    if "RAD21_MOUSE" in header:
        # Split the header by '|' and extract the gi ID
        gi_id = header.split('|')[1]  # The second element is the gi ID
        rad21_mouse_id = gi_id
        rad21_mouse_sequence = seq
        break

# Print the gi ID and sequence if found
if rad21_mouse_id and rad21_mouse_sequence:
    print("RAD21_MOUSE gi ID: {}".format(rad21_mouse_id))
    print("RAD21_MOUSE Sequence:\n{}".format(rad21_mouse_sequence))
else:
    print("RAD21_MOUSE gene not found.")

```

```python
# File path for the output FASTA file
output_file_path = "/Users/mohanavenkataphaneendrareddyalla/Desktop/Adv_pro_molecular_bio/Homework_3/New_RAD21_MOUSE.txt"

# Write the gi ID and gene sequence to the file in FASTA format
with open(output_file_path, "w") as output_file:
    # Write the header using format method
    output_file.write(">gi|{}|RAD21_MOUSE\n".format(rad21_mouse_id))
    # Write the sequence using format method
    output_file.write("{}\n".format(rad21_mouse_sequence))

print("RAD21_MOUSE sequence written to {}".format(output_file_path))
```
