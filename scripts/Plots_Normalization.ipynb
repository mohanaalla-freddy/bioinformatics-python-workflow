```python
import numpy as np
import matplotlib.pyplot as plt

def read_file(file_path):
    """Reads a CSV file and returns the data as a NumPy array."""
    return np.genfromtxt(file_path, delimiter=',')

def plot_histograms(data, output_file):
    """Plots histograms for the top 4 rows and saves the plot."""
    plt.figure(figsize=(12, 10))
    for i in range(4):
        plt.subplot(2, 2, i + 1)
        plt.hist(data[i], bins=8, color='blue', alpha=0.7)
        plt.title(f"Histogram for Row {i+1}")
        plt.xlabel('Values')
        plt.ylabel('Frequency')
    plt.tight_layout()
    plt.savefig(output_file)
    plt.show()

def plot_box_plots(data, output_file):
    """Plots box plots for the top 4 rows and saves the plot."""
    plt.figure(figsize=(8, 6))
    plt.boxplot(data[:4].T)
    plt.xticks([1, 2, 3, 4], [f"Row {i+1}" for i in range(4)])
    plt.title('Box Plots of Top 4 Rows')
    plt.ylabel('Values')
    plt.savefig(output_file)
    plt.show()

def normalize_data(data, output_file):
    """Normalizes the data by dividing each element by the sum of its column and saves it."""
    normalized_data = data / data.sum(axis=0)
    np.savetxt(output_file, normalized_data, fmt='%.6f', delimiter='\t')

def main():

    file_path = '/Users/mohanavenkataphaneendrareddyalla/Downloads/GSE152632_TRAPedFCNF_25Genes.csv'
    hist_output = '/Users/mohanavenkataphaneendrareddyalla/Downloads/Four_Genes_Histograms.pdf'
    boxplot_output = '/Users/mohanavenkataphaneendrareddyalla/Downloads/App_Syn1_BoxPlot.pdf'
    normalized_output = '/Users/mohanavenkataphaneendrareddyalla/Downloads/GSE152632_TRAPedFCNF_25Genes_Normalized.txt'
    
    # Load data
    data = read_file(file_path)
    
    # Generating the plots and saveing normalized data
    plot_histograms(data, hist_output)
    plot_box_plots(data, boxplot_output)
    normalize_data(data, normalized_output)
    print("Processing complete. Outputs saved.")

# Run the main function
if __name__ == "__main__":
    main()
```
